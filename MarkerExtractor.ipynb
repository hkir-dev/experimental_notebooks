# Marker Gene DB Template Extractor

Extracts marker gene db ROBOT template from the Human brain atlas all non-neuronal cells anndata (https://cellxgene.cziscience.com/collections/283d65eb-dd53-496d-adb7-7570c7caa443).

```python
import sys
!{sys.executable} -m pip install pandas
!{sys.executable} -m pip install anndata

import os
import pandas as pd
import anndata as ad
```

```python
nn_ad = ad.read_h5ad("data/human_all_non_neuronal.h5ad", backed="r")
print(nn_ad.obs.shape)
nn_ad.obs.head(4)
```

```python
print(nn_ad.var.shape)
nn_ad.var.head()
```

```python
ensembl_ids = nn_ad.var.index.unique().tolist()
records = list()
records.append(['ID', 'SC %', 'A rdfs:label', 'A oboInOwl:hasExactSynonym SPLIT=|'])
for ensembl_id in ensembl_ids:
    data = ['ensembl:' + ensembl_id, 'SO:0000704', nn_ad.var.loc[ensembl_id]['Gene'], '']
    records.append(data)

df = pd.DataFrame(records, columns=["ID", "TYPE", "NAME", "SYNONYMS"])

df.head()
```

```python
df.to_csv('data/ensg.tsv', sep="\t")
```
