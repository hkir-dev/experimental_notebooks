Add a new `cell_type` column to be compatible with the [CxG schema](https://github.com/chanzuckerberg/single-cell-curation/blob/main/schema/4.0.0/schema.md#cell_type)

```python
import ssl
from SPARQLWrapper import SPARQLWrapper, JSON
from cachetools import TTLCache, cached

cache = TTLCache(maxsize=256, ttl=3600)
ssl._create_default_https_context = ssl._create_unverified_context

@cached(cache)
def get_ontology_label(term_curie: str) -> str:
    term_id = term_curie.strip().replace(":", "_")
    sparql = SPARQLWrapper("https://ubergraph.apps.renci.org/sparql")
    query = """
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX obo: <http://purl.obolibrary.org/obo/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    SELECT DISTINCT *
    WHERE {{
      obo:{term_id} rdfs:label ?label .
    }}
    LIMIT 1
    """.format(term_id=term_id)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()

    bindings = results.get("results", {}).get("bindings", [])
    if not bindings:
        raise Exception("Ontology term not found: {}".format(term_curie))

    label = bindings[0].get("label", {}).get("value")
    return label

ann_data.obs['cell_type'] = ann_data.obs['cell_type_ontology_term_id'].apply(get_ontology_label)
```
